```python
import numpy as np
import cv2
import glob
import matplotlib.pyplot as plt
```

```python
img_name = "RPf_00152_obj.png"
img = cv2.imread("RPf_00152_obj.png",cv2.IMREAD_GRAYSCALE)
ax = plt.subplot()
ax.imshow(img,cmap="gray")
```

```python
gaus_tau1 = 0.75
gaussian_img = cv2.GaussianBlur(img,(41,41),0)
gaussian_edge_map = cv2.Canny(gaussian_img,threshold1=gaus_tau1,threshold2=2*gaus_tau1)

median_img = cv2.medianBlur(img,3)
median_tau1=0.75
median_edge_map = cv2.Canny(median_img,threshold1=median_tau1,threshold2=median_tau1*2)


prewitt_y = np.array([
    [1,1,1],
    [0,0,0],
    [-1,-1,-1]
])

prewitt_x = np.array([
    [-1,0,1],
    [-1,0,1],
    [-1,0,1]
])

smoothed_prewitt_x = cv2.filter2D(gaussian_img,cv2.CV_64F,prewitt_x)
smoothed_prewitt_y = cv2.filter2D(gaussian_img,cv2.CV_64F,prewitt_y)
smoothed_prewitt_y = 1/(1+np.abs(smoothed_prewitt_y))
smoothed_prewitt_x = 1/(1+np.abs(smoothed_prewitt_x))

_,axes = plt.subplots(3,2,figsize=(20,20))
for ax,image,title in zip(axes.flatten(),
                          [gaussian_img,gaussian_edge_map,
                           median_img,median_edge_map,smoothed_prewitt_x,smoothed_prewitt_y],
                          ["Gaussian Smoothed","GaussianCanny","Median Smoothed","Median Canny","Prewitt X","Prewitt Y"]):
    ax.imshow(image,cmap="gray")
    ax.set_title(title)
```

```python
img_name = "fourier_RPf_00152_obj.png"
img = cv2.imread(img_name,cv2.IMREAD_GRAYSCALE)
ax = plt.subplot()
ax.imshow(img,cmap="gray")
```

```python
gaus_tau1 = 3
gaussian_img = cv2.GaussianBlur(img,(41,31),0)
gaussian_edge_map = cv2.Canny(gaussian_img,threshold1=gaus_tau1,threshold2=3*gaus_tau1)
_,axes = plt.subplots(1,2,figsize=(30,30))

for ax,image,title in zip(axes.flatten(),
                          [gaussian_img,gaussian_edge_map,],
                          ["Gaussian Smoothed","GaussianCanny"]):
    ax.imshow(image,cmap="gray")
    ax.set_title(title)
```

```python
median_img = cv2.medianBlur(img,3)
median_tau1=0.75
median_edge_map = cv2.Canny(median_img,threshold1=median_tau1,threshold2=median_tau1*3)


prewitt_y = np.array([
    [1,1,1],
    [0,0,0],
    [-1,-1,-1]
])

prewitt_x = np.array([
    [-1,0,1],
    [-1,0,1],
    [-1,0,1]
])

smoothed_prewitt_x = cv2.filter2D(gaussian_img,cv2.CV_64F,prewitt_x)
smoothed_prewitt_y = cv2.filter2D(gaussian_img,cv2.CV_64F,prewitt_y)
smoothed_prewitt_y = 1/(1+np.abs(smoothed_prewitt_y))
smoothed_prewitt_x = 1/(1+np.abs(smoothed_prewitt_x))

_,axes = plt.subplots(3,2,figsize=(30,30))
for ax,image,title in zip(axes.flatten(),
                          [gaussian_img,gaussian_edge_map,
                           median_img,median_edge_map,smoothed_prewitt_x,smoothed_prewitt_y],
                          ["Gaussian Smoothed","GaussianCanny","Median Smoothed","Median Canny","Prewitt X","Prewitt Y"]):
    ax.imshow(image,cmap="gray")
    ax.set_title(title)
```

```python
img_name = "fourier_RPf_00102.png"
img = cv2.imread(img_name,cv2.IMREAD_GRAYSCALE)
ax = plt.subplot()
ax.imshow(img,cmap="gray")
```

```python
gaus_tau1 = 3
gaussian_img = cv2.GaussianBlur(img,(41,31),0)
gaussian_edge_map = cv2.Canny(gaussian_img,threshold1=gaus_tau1,threshold2=3*gaus_tau1)
_,axes = plt.subplots(1,2,figsize=(30,30))

for ax,image,title in zip(axes.flatten(),
                          [gaussian_img,gaussian_edge_map,],
                          ["Gaussian Smoothed","GaussianCanny"]):
    ax.imshow(image,cmap="gray")
    ax.set_title(title)
```

```python
size_mean = 80
mean_kernel = np.ones((size_mean,size_mean),np.float32)/(size_mean**2)
mean_img = cv2.filter2D(img,-1,mean_kernel)
mean_gauss_img = cv2.GaussianBlur(img,(21,21),0)

mean_edge_map = cv2.Canny(mean_img,threshold1=3,threshold2=3*3)
mean_gauss_img_edge_map = cv2.Canny(mean_gauss_img,threshold1=3,threshold2=3*3)

_,axes = plt.subplots(3,1,figsize=(30,30))

for ax,image,title in zip(axes.flatten(),
                          [mean_img,mean_edge_map,mean_gauss_img_edge_map],
                          ["Mean Smoothed","MeanCanny","mean_gauss_img_edge_map"]):
    ax.imshow(image,cmap="gray")
    ax.set_title(title)
```
