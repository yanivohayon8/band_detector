```python
import pandas as pd
import numpy as np
from sklearn import preprocessing
from scipy.cluster.hierarchy import dendrogram, linkage
from matplotlib import pyplot as plt
```

```python
df = pd.read_csv("1237.csv",index_col=0)
#df.columns = df.columns[1:]
df.head()
```

```python
df.r = df.r/255.0
df.g = df.g/255.0
df.b = df.b/255.0

df.width = df.width.astype(float)
df.width = df.width/df.width.sum()
df.head(50)
```

```python
X = df[df.columns[1:]]
X
```

```python
Z = linkage(X, 'ward',optimal_ordering=True)
fig = plt.figure(figsize=(25, 10))
dn = dendrogram(Z,labels=df.fragment.values.tolist())
plt.show()
```
